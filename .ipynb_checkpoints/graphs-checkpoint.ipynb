```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from matplotlib.pyplot import figure

%matplotlib inline
```

```python
filepath = '../Data Science HW/biodiversity-in-national-parks.csv'
df = pd.read_csv(filepath)
df.head()
```

```python
park_list = []

for index, row in df.iterrows():
    if row["Park Code"] not in park_list:
        park_list.append(row["Park Code"])

        
```

```python
# Part 1

endanger_spec_count = []

for i in range(0, len(park_list)):
    endanger_spec_count.append(0)
    

for index, row in df.iterrows():
    for j in range(0, len(park_list)):
        if row["Park Code"] == park_list[j] and row["Conservation Status"] == "Endangered":
            endanger_spec_count[j] = endanger_spec_count[j] + 1

            
```

```python
plt.figure(figsize=(35, 5))
plt.bar(park_list, endanger_spec_count, color='green')
plt.xlabel('Park Name (Code)')
plt.ylabel('# Of Endangered Species')
plt.savefig('graph1.png')
plt.show()
```

```python
# Part 2

mammal_count = []

for i in range(0, len(park_list)):
    mammal_count.append(0) 

for index, row in df.iterrows():
    for j in range(0, len(park_list)):
        if row["Park Code"] == park_list[j] and row["Category"] == "Mammal":
            mammal_count[j] = mammal_count[j] + 1


        
```

```python
plt.figure(figsize=(35, 5))
plt.bar(park_list, mammal_count, color='red')
plt.xlabel('Park Name (Code)')
plt.ylabel('# Of Mammals')
plt.savefig('graph2.png')
plt.show()
```

```python
filepath = '../Data Science HW/Data Analysis Experiment Data Collection.csv'
df = pd.read_csv(filepath)
df = df.dropna()
df.head()
```

```python
a_1 = []
a_2 = []
a_3 = []
a_effective = []
a_useful = []
a_fun = []

b_1 = []
b_2 = []
b_3 = []
b_effective = []
b_useful = []
b_fun = []

for index, row in df.iterrows():
        
    if "Boxes" in row["Which Prototype are you evaluating?"]:
        a_1.append(row["What was your time to complete task 1 (in seconds)?"])
        if row["What was your time to complete task 2 (in seconds)?"] > 1000:
            print("christ")
        else:
            a_2.append(row["What was your time to complete task 2 (in seconds)?"])
        a_3.append(row["What was your time to complete task 3 (in seconds)?"])
        a_effective.append(row["Overall, I thought this design was effective."])
        a_useful.append(row["Overall, I thought this design was useful."])
        a_fun.append(row["Overall, I thought this design was fun."])
        
    else:
        b_1.append(row["What was your time to complete task 1 (in seconds)?"])
        b_2.append(row["What was your time to complete task 2 (in seconds)?"])
        b_3.append(row["What was your time to complete task 3 (in seconds)?"])
        b_effective.append(row["Overall, I thought this design was effective."])
        b_useful.append(row["Overall, I thought this design was useful."])
        b_fun.append(row["Overall, I thought this design was fun."])


# Reject the null hypothesis (no difference; both equal) if t <= cv
# Other hypothesis:  1 < 2
# 1 = a_1
# 2 = b_1


cv = -1.645 # 90 cv
std = np.sqrt(((len(a_1) - 1) * pow(np.std(a_1),2) + (len(b_1) - 1) * pow(np.std(b_1), 2)) / (len(a_1) + len(b_1) - 2))
z = (np.mean(a_1) - np.mean(b_1)) / (std * np.sqrt(1/(len(a_1)) + 1/(len(b_1))))

if z <= cv:
    print("z <= cv. We reject the null hypothesis that the both of them are the same.")
else:
    print("z > cv. We do not reject the null hypothesis.")
    

# Test 2
    
cv = -1.645 # 90 cv
std = np.sqrt(((len(a_2) - 1) * pow(np.std(a_2),2) + (len(b_2) - 1) * pow(np.std(b_2), 2)) / (len(a_2) + len(b_2) - 2))
z = (np.mean(a_2) - np.mean(b_2)) / (std * np.sqrt(1/(len(a_2)) + 1/(len(b_2))))

if z <= cv:
    print("z <= cv. We reject the null hypothesis that the both of them are the same.")
else:
    print("z > cv. We do not reject the null hypothesis.")
    
# Test 3
    
cv = -1.645 # 90 cv
std = np.sqrt(((len(a_3) - 1) * pow(np.std(a_3),2) + (len(b_3) - 1) * pow(np.std(b_3), 2)) / (len(a_3) + len(b_3) - 2))
z = (np.mean(a_3) - np.mean(b_3)) / (std * np.sqrt(1/(len(a_3)) + 1/(len(b_3))))

if z <= cv:
    print("z <= cv. We reject the null hypothesis that the both of them are the same.")
else:
    print("z > cv. We do not reject the null hypothesis.")
    
print("Mean of A is effective scores: " + str(np.mean(a_effective)))
print("Mean of A is useful scores: " + str(np.mean(a_useful)))
print("Mean of A is fun scores: " + str(np.mean(a_fun)))

print("Mean of B is effective scores: " + str(np.mean(b_effective)))
print("Mean of B is useful scores: " + str(np.mean(b_useful)))
print("Mean of B is fun scores: " + str(np.mean(b_fun)))
```
